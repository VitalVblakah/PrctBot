```
def generate_message(messages):
    chat_completion = openai.ChatCompletion.create(model="gpt-3.5-turbo", messages=messages)
    return chat_completion["choices"][0]["message"]["content"]

first_description = generate_message([
    {
        "role": "user",
        "content": "You are a dungeon master. Describe start location for campaign in fantasy setting in 50 words."
    }
])
print(first_description)

messages = [
    {
        "role": "system",
        "content": "You are a dungeon master. User will day actions and you as game master need to ract to the. For example if player talks to npc you need to roleplay."
    },
    {
        "role":"assistant",
        "content":first_description
    }
]
while True:
    command = input("Your command: ")
    messages.append({
        "role":"user",
        "content":command
    })
    reaction = generate_message(messages)
    messages.append({
        "role":"assistant",
        "content":reaction
    })

    print(f"Your command: {command}")
    print(reaction)
```
